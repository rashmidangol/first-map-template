```python
#import py libaries
import pandas as pd
import hvplot.pandas
import holoviews as hv
import osmnx as osm
```

```python
tamu_boundary = osm.geocode_to_gdf("R8556620", by_osmid=True)
tamu_boundary
```

```python
tamu_boundary.plot()
```

```python
#make interactive map
tamu_map = tamu_boundary.hvplot(geo=True, tiles="EsriImagery",
                                # Change the colors
                                    fill_color='white', fill_alpha=0.2,
                                    line_color='skyblue', line_width=5,
                                    # Change the image size
                                    frame_width=400, frame_height=400)
tamu_map
```

```python

# Save the map as a file to put on the web
hv.save(tamu_map, 'tamu_map.html')

# Display the map
tamu_map
```
